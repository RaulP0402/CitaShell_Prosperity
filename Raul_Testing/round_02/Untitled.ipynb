```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/raullopez/Desktop/round2_data/prices_round_2_day_-1.csv', sep=';')
df.head()
```

```python
# Change all NaN values to 0
df.fillna(0, inplace=True)
# pd.isna(df).any()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize columns
scaler = MinMaxScaler()
columns_to_normalize = df.columns.difference(['timestamp'])  # Exclude 'Timestamp' column
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
```

```python
import seaborn as sns

sns.pairplot(df)
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['timestamp'], df['ORCHIDS'], label='ORCHIDS')
plt.plot(df['timestamp'], df['TRANSPORT_FEES'], label='TRANSPORT_FEES')
plt.plot(df['timestamp'], df['EXPORT_TARIFF'], label='EXPORT_TARIFF')
plt.plot(df['timestamp'], df['IMPORT_TARIFF'], label='IMPORT_TARIFF')
plt.plot(df['timestamp'], df['SUNLIGHT'], label='SUNLIGHT')
plt.plot(df['timestamp'], df['HUMIDITY'], label='HUMIDITY')
plt.xlabel('Timestamp')
plt.ylabel('Normalized Values')
plt.title('Normalized Plot of Columns Over Time')
plt.legend()
plt.show()
```
